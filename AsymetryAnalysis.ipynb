# Asymmetry Analysis: 

First, we need to read the data in off the hard drive, turn it into a dataframe and extract important information from filenames. This will then be used to plot time series data, extract means, and eventually export to .csv for plotting using more elegant methods. 

```python
import sys
sys.path.append("/Users/joseph/Code/Python")

import numpy as np
import pandas as pd
import glob as glob
import forJake as fj
import joeHelper as jh
import matplotlib.pyplot as plt
#Find all related datafiles
framerate = 15
datafiles = glob.glob('/Users/joseph/Dropbox/OFExperiments/Asymetry/DR*YL0050XXX/DR*YL0050XXX.csv')


#Create pandas dataframe for each file
pandaData = [pd.read_csv(fname, names=['frame', 'n','theta','x','y','deg']) for fname in datafiles]

#Calculate the time for each frame
for df in pandaData:
    df['time'] = df['frame']/ framerate

#Make a list of ID's, Regex would probably be prettier... but this works 
file_ID = [file[47:49] for file in datafiles]

#Make a list of FileNames, for use in plotting or otherwise. 
#Make a list of DV's 
file_DV = [file[48:49] for file in datafiles]
```

## Response to onset of stimulus

Here we select only the frames corresponding to +/- 3 seconds of the stimulus onset(at 60s). We'll look at it in terms of the angle vs. time

### Angle vs. Time: Control Fish

```python
#Slice the frames of interest
Angs = [df.deg[(df.frame > 855) & (df.frame <=945)] for df in pandaData]
times = [df.time[(df.frame > 855) & (df.frame <=945)] for df in pandaData]
%matplotlib inline
for idx,Ang in enumerate(Angs):
    if file_ID[idx][0] == 'C':
        plt.figure()
        plt.title(file_ID[idx])
        plt.xlabel('Time')
        plt.ylabel('Angle (deg)')
        plt.plot(times[idx], Ang)
        plt.plot([60,60], [-180, 180], color='k', linestyle='--', linewidth=2)



```
