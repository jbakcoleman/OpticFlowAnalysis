## RI Analysis Minute by Minute

```python
#Make a list of RI's for the first half. 
before_RI = [fj.getRI(df.deg[df.frame < 900]) for df in pandaData]
#Make a list of RI's for the second half 
after_RI = [fj.getRI(df.deg[(df.frame > 900) & (df.frame <1800)]) for df in pandaData]
#Make a new dataframe
d = {'fileID' : file_ID, 'fileDV' : file_DV, 'beforeRI' : before_RI, 'afterRI' : after_RI}
output = pd.DataFrame(d)
#descriptive statistics
output['diff'] = output.afterRI - output.beforeRI
output['Treatment'] = 'Control' 
output.Treatment[output.fileDV !='C'] = 'Asymmetric'
```

```python

#Slice control and treatment data
control = output[output['fileDV']=='C']
treatment = output[output['fileDV']!='C' ]

np.std(treatment['diff'])
np.std(control['diff'])


#Calculate average RI before and after, as well as SE
before = (np.mean(control['beforeRI']), np.mean(treatment['beforeRI']))
after = (np.mean(control['afterRI']), np.mean(treatment['afterRI']))
beforeSE = (np.std(control['beforeRI']) / np.sqrt(8), np.std(treatment['beforeRI']) / np.sqrt(8))
afterSE = (np.std(control['afterRI']) / np.sqrt(8), np.std(treatment['afterRI']) / np.sqrt(8))


#Plot
fig, ax = plt.subplots()
width = 0.35
rects1 = ax.bar(np.arange(2), before, width,color ='grey')
ax.errorbar(np.arange(2) +  width / 2, before, yerr=beforeSE, color='black',fmt = '|')

rects2 = ax.bar(np.arange(2)+width, after, width,color ='white', hatch='/')
ax.errorbar(np.arange(2) +  width / 2 + width, after, yerr=afterSE, color='black', fmt = '|')


plt.rcParams.update({'font.size': 22})

ax.set_ylabel('RI')
ax.set_title('One minute')
ax.set_xticks(np.arange(2)+width)
ax.set_xticklabels(('Control', 'Asymmetric') )
ax.axis([-.1, 1.8, 0, 1.0])
ax.legend( (rects1[0], rects2[0]), ('Pre-stimulus', 'Stimulus') )
print('t, p, df=12')
print(ttest_ind(control['diff'], treatment['diff']))


fig.savefig('/Users/joseph/Dropbox/OFExperiments/Paper/60SecondRI.pdf', format='pdf')
```

### Angular velocity 3 seconds

```python
def angVel(array):
    '''
    Returns RI value for a given numpy array
    assumes -180 to 180
    '''
    array = np.diff(array)
    array[array > 180] -= 180
    array[array < -180] += 180
    
    return array

def avgAng(array):
    #Convert to radians and unit vector
    array = array / 180
    array = array * np.pi
    x = np.sin(array)
    y = np.cos(array)
    #Average each component
    totalx = np.average(x)
    totaly = np.average(y)
    
    #Convert back to angle
    return np.arctan2(totalx, totaly) * 180 / np.pi


#Make a list of Angular velocities's for the first half. 
before_AV = [avgAng(angVel(df.deg[(df.frame >= 855) & (df.frame <=900)])) for df in pandaData]
#Make a list of RI's for the second half 
after_AV = [avgAng(angVel(df.deg[(df.frame >= 900) & (df.frame <=945)])) for df in pandaData]
                       
#Make a new dataframe
d = {'fileID' : file_ID, 'fileDV' : file_DV, 'beforeAV' : before_AV, 'afterAV' : after_AV}
outputAV = pd.DataFrame(d)

#Calculate the change in angular velocity
outputAV['diff'] = outputAV.beforeAV- outputAV.afterAV
outputAV['Treatment'] = 'Control' 
outputAV.Treatment[output3.fileDV !='C'] = 'Asymmetric'

#Flip R's to L's so they're analagous. 
outputAV.afterAV[output3.fileDV == 'R'] = outputAV.afterAV[output3.fileDV == 'R'] * -1
outputAV.beforeAV[output3.fileDV == 'R'] = outputAV.beforeAV[output3.fileDV == 'R'] * -1

#Change controls,c onsistent with changing treatments
outputAV.afterAV[output3.fileID == 'CA'] = outputAV.afterAV[output3.fileID == 'CA'] * -1
outputAV.beforeAV[output3.fileID == 'CA'] = outputAV.beforeAV[output3.fileID == 'CA'] * -1
outputAV.afterAV[output3.fileID == 'CB'] = outputAV.afterAV[output3.fileID == 'CB'] * -1
outputAV.beforeAV[output3.fileID == 'CB'] = outputAV.beforeAV[output3.fileID == 'CB'] * -1
outputAV.afterAV[output3.fileID == 'CC'] = outputAV.afterAV[output3.fileID == 'CC'] * -1
outputAV.beforeAV[output3.fileID == 'CC'] = outputAV.beforeAV[output3.fileID == 'CC'] * -1
outputAV.afterAV[output3.fileID == 'CD'] = outputAV.afterAV[output3.fileID == 'CD'] * -1
outputAV.beforeAV[output3.fileID == 'CD'] = outputAV.beforeAV[output3.fileID == 'CD'] * -1


outputAV
```
